## Aide pour le merge

Pour "merger" deux dataframe, il faut utiliser la fonction _merge_ et dire à la fonction quelles sont les colonnes qui doivent être utilisées pour lier les lignes des DataFrame. 

Par exemple ici, on veut ajouter sur chaque ligne de q3anim la population du pays correspondant à cette ligne. On va donc dire à _merge_ de chercher sur chaque ligne de q3anim le code du pays de la ligne, et de trouver dans datapays la ligne avec le même code pays, afin d'ajouter les infos de cette ligne-là dans q3anim.

Ainsi, les paramètres _left_on_ et _right_on_ doivent valoir les noms des colonnes contenant les codes des pays pour chacune des dataframe. Pour trouver le nom de ces colonnes, j'affiche dans les deux cellules suivantes les colonnes des deux dataframes.

```python
datapaysdispo.columns
```

```python
q3anim.columns
```

```python
# fusionnons  les deux  tables pour repondre a  la question  pour Anim 
q3anim_pop = q3anim.merge(datapaysdispo, left_on = 'Code zone' , right_on = 'Code zone')
q3anim_pop
```
